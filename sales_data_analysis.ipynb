Sales Data Analysis – Key Findings

After analyzing the dataset, several important patterns became clear.

First, not all sales are actually profitable. Around 19% of transactions result in a loss, which is quite significant. This suggests that the company is not consistently making effective pricing decisions.

One of the main reasons for these losses appears to be high discount levels. As the discount increases, the average profit decreases. In fact, once the discount goes above approximately 30%, transactions tend to become unprofitable. The largest losses occur at a 50% discount level.

Another important observation is related to product categories. While some products generate strong profits (such as copiers and phones), others consistently perform poorly. In particular, the "Tables" sub-category stands out as the biggest source of losses.

Overall, the analysis shows that profitability is strongly influenced by both discount strategy and product selection. Reducing excessive discounts and focusing on high-performing product categories could significantly improve business results.
